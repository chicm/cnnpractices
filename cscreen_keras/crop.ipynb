```python
import os, shutil, glob, sys
import pandas as pd
import numpy as np
import cv2

DATA_DIR = '/home/chicm/ml/cnnpractices/cervc/data/full'
TRAIN_DIR = DATA_DIR+'/train'
TEST_DIR = DATA_DIR + '/test'
VALID_DIR = DATA_DIR + '/valid'

TGT_DIR = '/home/chicm/ml/cnnpractices/cervc/data/crop'
TGT_TRAIN = TGT_DIR+'/train'
TGT_TEST = TGT_DIR + '/test'
TGT_VAL = TGT_DIR + '/valid'
```

```python
train_df = pd.read_csv('/home/chicm/ml/cnnpractices/cervc/bbox_train.csv')
train_df.head()
```

```python
#for i in range(train_df.shape[0]):
#    if train_df['sh0_start'][i] + 50 >  train_df['sh0_end'][i] or train_df['sh1_start'][i] + 50 > train_df['sh1_end'][i]:
#        print(train_df.values[i])
    
```

```python
for i in range(train_df.shape[0]):
    f = train_df['image_name'][i]
    cls = train_df['clss'][i]
    if i < 6642:
        continue
    fname = TRAIN_DIR + '/Type_' +str(cls+1) + '/' + f
    tgtfn = TGT_TRAIN + '/Type_' +str(cls+1) + '/' + f
    print(fname, i)
    img = cv2.imread(fname)
    res = cv2.resize(img, (train_df['img_shp_0'][i], train_df['img_shp_1'][i]))
    crop = res[train_df['sh0_start'][i]:train_df['sh0_end'][i], train_df['sh1_start'][i]:train_df['sh1_end'][i]]
    res2 = cv2.resize(crop, (224, 224))
    cv2.imwrite(tgtfn, res2)
```

```python
val_df = pd.read_csv('/home/chicm/ml/cnnpractices/cervc/bbox_val.csv')
val_df.head()
```
